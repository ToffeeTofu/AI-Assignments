```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import random
import pandas as pd

data = pd.read_csv('student_performance_dataset.csv')
#Store each of the columns in their respective variable
x = data['Attendance_Rate']
y = data['Past_Exam_Scores']
m = x.shape[0]

# Plots graph
plt.scatter(x, y, color="blue", label="Actual data")
plt.xlabel("Attendance")
plt.ylabel("Grade")

theta0 = random.uniform(0.0,100.0)
theta1 = random.uniform(0.0,1.0)
init_theta0 = theta0
init_theta1 = theta1

alpha = 0.00025
counter = 0
y_hat = theta0 + theta1 * x          # fill with a vectorized expression using theta0, theta1, x
y_pred = y_hat
errors = y_hat - y
plt.plot(x, y_pred, color=(random.randint(0,255)/255,random.randint(0,255)/255,random.randint(0,255)/255), label="First Best Fit Line")
J_1 = np.dot (errors, errors) / (2*m) # same as np.dot(errors, errors)
counter = 0

for i in range(10000):
  grad_theta0 = 1/m * sum(errors)                 # (1/m) * sum(errors)
  grad_theta1 = 1/m * sum(errors*x)                 # (1/m) * sum(errors * x)

  theta0 = theta0 - alpha * grad_theta0
  theta1 = theta1 - alpha * grad_theta1

  y_hat = theta0 + theta1 * x          # fill with a vectorized expression using theta0, theta1, x
  y_pred = y_hat
  errors = y_hat - y
  counter = counter + 1
  if(counter == 500):
    plt.plot(x, y_pred, color=(random.randint(0,255)/255,random.randint(0,255)/255,random.randint(0,255)/255), label="Best fit line")
    counter = 0

J_2 = np.dot (errors, errors) / (2*m) # same as np.dot(errors, errors)
print("***Statistics***")
print("First Cost Function: ")
print(J_1)
print("Inital theta0: ")
print(init_theta0)
print("Inital theta1: ")
print(init_theta1)
print("Learning Rate: ")
print(alpha)
print("Final theta0: ")
print(theta0)
print("Final theta1: ")
print(theta1)
print("Final Cost Function: ")
print(J_2)

#shows graph
plt.tight_layout()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
